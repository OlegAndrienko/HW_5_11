```python
import os
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')

#Створюю дата сет для аналізу
path = '/content/gdrive/MyDrive/HW_5_data'
# функція для отримання списку потрібних тек з даними
def get_folder_(path): #
    folders = []
    if os.path.isdir(path):
        folders = os.listdir(path)
        return folders
    else:
        return f'"{path}" is not a directory'

list_of_folders = get_folder_(path)
print(f'get_folder_(path) = {get_folder_(path)}')


#create function to add data from files to data frame
def add_data_to_df(path):
    df = pd.DataFrame() # create empty data frame
    # add data from all folders to data frame
    for folder in list_of_folders:
        folder_path = path + '/' + folder
        for file in os.listdir(folder_path):
            data_file = pd.read_csv(folder_path + '/' + file)
            data_file['Goal'] = folder
            df = pd.concat([df, data_file])
    return df

df = add_data_to_df(path)
#displey first 5 rows
print(f'df.head() = \n {df.head()}')

```

# Новий розділ

```python
print(f'df.shape = {df.shape}')
#create list of columns names
columns = df.columns.tolist()
print(f'columns = {columns}')
```

```python
#devide features and target
X = df[columns[:-1]]
y = df[columns[-1]]
```

```python
#split data to train and test
from sklearn. model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
#check shape of train and test data
print(f'X_train.shape = {X_train.shape}')
print(f'X_test.shape = {X_test.shape}')
print(f'y_train.shape = {y_train.shape}')
print(f'y_test.shape = {y_test.shape}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
#LEARNING svm
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
```

```python
#learning random forest
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
#forecast svm
y_pred_svm = svm.predict(X_test)
#forecast random forest
y_pred_rfc = rfc.predict(X_test)
```

```python
#check accuracy
print(f'svm.score(X_test, y_test) = {svm.score(X_test, y_test)}')
print(f'rfc.score(X_test, y_test) = {rfc.score(X_test, y_test)}')
```
